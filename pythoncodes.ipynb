```python
import numpy as np
import pandas as pd
```

```python
M = np.pi/180.0*np.array([15,
30,
45,
60,
75,
90,
105,
120,
135,
150,
165])

Mpi = np.array([15,
30,
45,
60,
75,
90,
105,
120,
135,
150,
165])

Mstr = ["15","30","45","60","75","90","105","120","135","150","165"]

e = np.array([0.239653,
              0.480475,
              0.723624,
              0.970224,
              1.22134,
              1.4779,
              1.74064,
              2.00995,
              2.28574,
              2.56731,
              2.85326])

ecc = np.array([0.20563,0.00677,0.09341,0.04839,0.05415,0.04716,0.00858])

r_1 = {}
theta_1 = {}
r_2 = {}
theta_2 = {}
r_3 = {}
theta_3 = {}

difr_12 = {}
difr_13 = {}
difa_12 = {}
difa_13 = {}

data = {}

datatoexcel = {}


planet = ["Mercury","Venus","Mars","Jupiter","Saturn","Uranus", "Neptune"]
```

```python
for i in range(len(planet)):
    r_1[planet[i]] = 1 + ecc[i]*np.cos(e)
    theta_1[planet[i]] = np.arccos((ecc[i] + np.cos(e))/(1+ecc[i]*np.cos(e)))*(180.0/np.pi)

r_1, theta_1
```

```python
for i in range(len(planet)):
    r_2[planet[i]] = np.sqrt(4*np.power(ecc[i],2) + ((1-ecc[i]**2)/(1+ecc[i]*np.cos(M)))**2 + 4 * ecc[i] * ((1-ecc[i]**2)/(1+ecc[i]*np.cos(M))) * np.cos(M))
    theta_2[planet[i]] = np.arccos((r_2[planet[i]]**2+4*ecc[i]**2-((1-ecc[i]**2)/(1+ecc[i]*np.cos(M)))**2)/(4*r_2[planet[i]]*ecc[i]))*(180.0/np.pi)

r_2, theta_2
```

```python
for i in range(len(planet)):
    middle = np.sqrt(1-ecc[i]**2*np.sin(M)**2)*np.cos(M)+ecc[i]*np.sin(M)**2
    r_3[planet[i]] = np.sqrt(ecc[i]**2 + 1 + 2*ecc[i]*middle)
    theta_3[planet[i]] = np.arccos((r_3[planet[i]]**2 + ecc[i]**2 - 1)/(2*r_3[planet[i]]*ecc[i]))*(180.0/np.pi)

r_3, theta_3
```

```python
for i in range(len(planet)):
    difr_12[planet[i]] = r_1[planet[i]] - r_2[planet[i]]
    difr_13[planet[i]] = r_1[planet[i]] - r_2[planet[i]]
    difa_12[planet[i]] = theta_1[planet[i]] - theta_2[planet[i]]
    difa_13[planet[i]] = theta_1[planet[i]] - theta_3[planet[i]]

difr_12, difr_13, difa_12, difa_13
```

```python
for i in range(len(planet)):
    data[i] = {"r_1":r_1[planet[i]], "r_2":r_2[planet[i]], "r_3":r_3[planet[i]], "theta_1":theta_1[planet[i]], "theta_2":theta_2[planet[i]], "theta_3":theta_3[planet[i]], "r_1-r_2":difr_12[planet[i]], "r_1-r_3":difr_13[planet[i]], "theta_1-theta_2":difa_12[planet[i]], "theta_1-theta_3":difa_13[planet[i]]}

data
```

```python
writer = pd.ExcelWriter('all4mars.xlsx')


for i in range(len(planet)):
    datatoexcel[i] = pd.DataFrame(data[i])  # header参数表示列的名称，index表示行的标签
    datatoexcel[i].index = Mpi
    datatoexcel[i].to_excel(writer, planet[i], float_format='%.8f')
# 注意写多个的时候要把所有数据都给了writer后再调用save函数，要不然可能只有一个sheet

writer.save()
writer.close()
```
